```python
import csv
import urllib.request
import json

# Set the API key
api_key = ''

# Set the video ID for which you want to retrieve comments
video_id = 'lh6Ltp5Ew9k'

# Retrieve the comments for the video
url = f'https://www.googleapis.com/youtube/v3/commentThreads?key={api_key}&textFormat=plainText&part=snippet&videoId={video_id}&maxResults=10000'
comments = []
while True:
    response = urllib.request.urlopen(url)
    data = json.loads(response.read().decode())
    for item in data['items']:
        comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
        comments.append([comment])  # Store each comment in a list to save to CSV

    # Check if there are more comments to retrieve
    if 'nextPageToken' in data:
        next_page_token = data['nextPageToken']
        url = f'https://www.googleapis.com/youtube/v3/commentThreads?key={api_key}&textFormat=plainText&part=snippet&videoId={video_id}&maxResults=10000&pageToken={next_page_token}'
    else:
        break

# Save the comments to a CSV file
csv_file = 'test.csv'
with open(csv_file, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(comments)

print(f'{len(comments)} comments have been saved to {csv_file}.')
```

```python
import csv
import pandas as pd

# Read the CSV and calculate lengths
with open('youtube_comments.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    lengths = [len(row[0]) for row in reader]

# Create a DataFrame
df = pd.DataFrame(lengths, columns=['Length'])

# Display the DataFrame
print(df)

# Optionally, display the number of comments
print(f'Total number of comments: {len(df)}')
```
